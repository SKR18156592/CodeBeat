```python
def fun1(x, y):
    m = 0
    for i in range(x):
        for j in range(y):
            m += i*j 
    return m
```

```python
def matrix_sum(x, y):
    total = 0
    for i in range(x):
        row = 0
        for j in range(y):
            row += i + j
        total += row
    return total

obj = code_tracker(matrix_sum)
obj(5, 6)  # Expect outer loop ~2x inner time
```

```python
def rec_mult_fib(x, y):
    def fib(n):
        if n <= 1:
            return n
        return fib(n-1) + fib(n-2)
    return fib(x) * y

obj = code_tracker(rec_mult_fib)
obj(8, 10)  # Recursive calls dominate time
```
