```python
from codebeat import code_tracker
```

```python
def fun1(x, y):
    m = 0
    for i in range(x):
        for j in range(y):
            m += i*j 
    return m
```

```python
obj = code_tracker(fun1) # modify, create executable
obj(3, 4) # for loop for n times and call fun and display final result
```

```python
def matrix_sum(x, y):
    total = 0
    for i in range(x):
        row = 0
        for j in range(y):
            row += i + j
        total += row
    return total

obj = code_tracker(matrix_sum)
obj(5, 6)  # Expect outer loop ~2x inner time
```

```python
def branch_counter(x, y):
    count = 0
    for i in range(x):
        if i < y:
            count += 1
        else:
            count *= 2
    return count

obj = code_tracker(branch_counter)
obj(10, 5)  # First half cheap, second expensive
```

```python
def math_series(x, y):
    s = 0.0
    for i in range(x):
        s += sin(i) * cos(y * i) * exp(-i / 10.0)
    return s * y

obj = code_tracker(math_series)
obj(20, 2)  # Trig/exp functions slower than arithmetic
```

```python
def list_builder(x, y):
    results = []
    for _ in range(x):
        lst = []
        for j in range(y):
            lst.append(j * 2)
        results.append(sum(lst))
    return sum(results)

obj = code_tracker(list_builder)
obj(4, 8)  # Append dominates in inner loop
```
